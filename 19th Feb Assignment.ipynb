```python
# 1.Ans.

Flask is a popular lightweight web framework for Python that is designed to 
make it easy to build web applications quickly and efficiently. It is known for its
simplicity and flexibility, making it a great choice for small to medium-sized projects
that require a fast development cycle.

    Some advantages of using Flask Framework are:

Minimalistic: Flask is a minimalistic framework that comes with only the basic features required to
build web applications. This makes it easy to learn and use, and allows developers to build 
applications quickly.

Flexible: Flask is a flexible framework that allows developers to choose the tools and 
libraries that they want to use. This means that developers can customize their applications 
to meet their specific needs.

Easy to use: Flask is designed to be easy to use and understand. Its simple syntax and
intuitive structure make it easy for developers to get started and build applications quickly.

Modular: Flask is a modular framework that allows developers to add or remove features as needed. 
This means that developers can start with a basic application and add features as they are needed,
without having to rewrite the entire application.

```

```python
# 2.Ans.

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
    
```

```python
# 3.Ans.

In Flask, app routing refers to the process of mapping URLs to functions or 
views that will handle requests made to those URLs. When a user makes a request to a particular URL
in a Flask application, the app routing system determines which function or view should be executed
to handle that request.

App routes are a key part of developing web applications in Flask. 
They allow developers to define the behavior of their application based on the 
URL that a user requests. App routes are useful for creating RESTful APIs, handling 
form submissions, and rendering dynamic content based on user input.
```

```python
# 4.Ans.

from flask import Flask

app = Flask(__name__)


@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '''Company Name: ABC Corporation
              Location: India
              Contact Detail: 999-999-9999'''
    

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
# 5.Ans.

In Flask, the url_for() function is used for URL building.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def hello():
    return 'Hello, World!'

with app.test_request_context():
    url = url_for('hello')
    print(url)
```
